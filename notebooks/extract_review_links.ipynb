```python
with open("../data/liens_revue.md", "r") as md_file:
    md_lines = md_file.readlines()
    
```

```python
last_date_value = None
links = []

for line in md_lines[6:30]:
    line_split = line.split("|")

    if line_split[1]=="":
        date = last_date_value
    else:
        date = line_split[1].strip()
        last_date_value = date
    
    date = "-".join(date.split("/")[::-1])
        
    url = line_split[2].strip()
        
    links.append({"link": url, "upload_date": date})
        
```

```python
links
```

```python
import json

with open("review_links_medium.json", "w") as json_file:
    json_file.write(json.dumps(links, indent=4))
```

```python
import csv
```

```python
fieldnames = ["content_path", "date", "already_read", "source"]

with open("review_links_large.csv", "w", newline='') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(links)
```
